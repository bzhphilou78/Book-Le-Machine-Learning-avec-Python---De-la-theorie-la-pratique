```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Data/exemple_simple_kmeans-3.csv', sep=',' )
plt.figure(1, figsize=(5, 5)) 
plt.scatter(data.X, data.Y,s=60)
```

```python
fig, axs = plt.subplots(4, 3)
inertie_lst = []
silhouette_lst = []
k=2 #Ici, la premire valeur du paramtre k est  gale  2.
for row in range(4):
    for col in range(3):
        print("k= ",k)
        model = KMeans(n_clusters=k)
        model.fit(data)    
        inertie_lst.append(model.inertia_)
        #Calcul du coefficient silhouette
        silhouette_lst.append(silhouette_score(data, model.labels_))        
        centers = model.cluster_centers_
        axs[row,col].set_title(f"k = {str(k)}, inertie = {round(model.inertia_,2)}")
        axs[row,col].scatter(centers[:,0], centers[:,1],c='red',s=100)
        axs[row,col].scatter(data.X, data.Y,c='#000000',s=1)
        for i in range(len(model.labels_)) :
            axs[row,col].text(data.X[i], data.Y[i], model.labels_[i])
        k+=1
        
fig.set_figheight(20)
fig.set_figwidth(15)  
              
```
